```python
import pandas as pd

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('ml-1m/users.dat', sep='::', header=None,
names=unames, engine='python')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('ml-1m/ratings.dat', sep='::', header=None,
names=rnames, engine='python')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('ml-1m/movies.dat', sep='::', header=None,
names=mnames, engine='python')

data = pd.merge(pd.merge(ratings, users), movies)

data[:10]
```

```python
store = pd.HDFStore('m1_1m_data.h5')
store['obj1'] = data
store
```

```python
import numpy as np
meanrating_by_title=data.groupby('title').rating.mean()
meanrating_by_title[:10]
```

```python
#How many movies have an average rating over 4.5 overall?
active_title=meanrating_by_title.index[meanrating_by_title>=4.5]
top_count=meanrating_by_title[active_title].count()
top_count
```

```python
#How many movies have an average rating over 4.5 among men? How about women?
mean_ratings=data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
men_mean_ratings=mean_ratings['M']
men_active_titls=men_mean_ratings.index[men_mean_ratings>=4.5]
men_top_count=men_mean_ratings.ix[men_active_titls].count()
men_top_count
```

```python
women_mean_ratings=mean_ratings['F']
women_active_titls=women_mean_ratings.index[women_mean_ratings>=4.5]
women_top_count=women_mean_ratings.ix[women_active_titls].count()
women_top_count
```

```python
#How many movies have an median rating over 4.5 among men over age 30? How about women over age 30?
data2=data[data.age>=30]
median_ratings_above30=data2.pivot_table('rating',index='title', columns='gender',aggfunc='median')
men_above30_median_ratings=median_ratings_above30['M']
men_above30_active_titls=men_above30_median_ratings.index[men_above30_median_ratings>=4.5]
men_above30_top_count=men_above30_median_ratings.ix[men_above30_active_titls].count()
men_above30_top_count
```

```python
women_above30_median_ratings=median_ratings_above30['F']
women_above30_active_titls=women_above30_median_ratings.index[women_above30_median_ratings>=4.5]
women_above30_top_count=women_above30_median_ratings.ix[women_above30_active_titls].count()
women_above30_top_count
```

```python
#Ten top rating times movies
ratingnum_by_title=data.groupby('title').rating.size()
top_ratingnum=ratingnum_by_title.sort_values(ascending=False)
top_ratingnum[:10]
```
